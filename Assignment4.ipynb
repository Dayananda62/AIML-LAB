```python
import matplotlib.pyplot as plt

x_values = []
y_values = []

with open('test.txt', 'r') as file:
    for line in file:
        x, y = map(float, line.split())
        x_values.append(x)
        y_values.append(y)

plt.plot(x_values, y_values, marker='o')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')

plt.show()
```

```python
import pandas as pd

user_input1=input("Enter values separated by commas: ")
user_input2=input("Enter values separated by commas: ")

ds11=[int(x) for x in user_input1.split(',')]
ds22=[int(x) for x in user_input2.split(',')]

ds1=pd.Series(ds11)
ds2=pd.Series(ds22)

ds = ds1 + ds2
print("Add two Series:")
print(ds)
print("Subtract two Series:")
ds = ds1 - ds2
print(ds)
print("Multiply two Series:")
ds = ds1 * ds2
print(ds)
print("Divide Series1 by Series2:")
ds = ds1 / ds2
print(ds)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
Y = np.array([1.5, 3.8, 6.7, 9.0, 11.2, 13.6, 16])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

mse = mean_squared_error(Y, Y_pred)

print("Linear Regression Analysis")
print("---------------------------")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")

print("\nPredicted Values:")
print("----------------")
for xi, yi_true, yi_pred in zip(X.flatten(), Y, Y_pred):
    print(f"X = {xi: .1f} -> True Y = {yi_true: .2f}, Predicted Y = {yi_pred: .2f}")
print("\nMean Squared Error(MSE):")

print(f"{mse:.2f}")
```
